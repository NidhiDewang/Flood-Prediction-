```python
PAST_DAYS = 120
```

```python
dataset_train = pd.read_csv('india flood dataset.csv')
training_set = dataset_train.iloc[:,14:15].values
print(training_set)
```

```python
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
training_set_scaled = sc.fit_transform(training_set)
print(training_set_scaled)
```

```python
df.head()
```

```python
# Creating a data structure with 60 timesteps and 1 output
X_train = []
y_train = []
```

```python
for i in range(PAST_DAYS, 4116):
    X_train.append(training_set_scaled[i-PAST_DAYS:i, 0])
    y_train.append(training_set_scaled[i, 0])
X_train, y_train = np.array(X_train), np.array(y_train)
print(X_train)
print(y_train)
print(len(X_train))
print(len(y_train))
```

```python
## Reshaping
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
#Building the RNN

# Importing the Keras libraries and packages
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
# Initialising the RNN
regressor = Sequential()
```

```python
# Adding the first LSTM layer and some Dropout regularisation
regressor.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 1)))
regressor.add(Dropout(0.2))
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))
regressor.add(LSTM(units = 50))
regressor.add(Dropout(0.2))

regressor.add(Dense(units = 1))

regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')
regressor.fit(X_train, y_train, epochs = 1, batch_size = 32)

```

```python
import io
df=pd.read_csv(io.BytesIO(uploaded['testdataset.csv']))
```

```python
dataset_test = pd.read_csv('testdataset.csv')
real_value = dataset_test.iloc[:, 14:15].values
print(real_value)
print(dataset_test)
```

```python
# Getting the predicted
dataset_total = pd.concat((dataset_train['ANNUAL'], dataset_test['ANNUAL']), axis = 0)
inputs = dataset_total[len(dataset_total) - len(dataset_test) - PAST_DAYS:].values
print(inputs)
inputs = inputs.reshape(-1,1)
inputs = sc.transform(inputs)
```

```python
X_test = []
for i in range(PAST_DAYS, PAST_DAYS+20):
    X_test.append(inputs[i-PAST_DAYS:i, 0])
X_test = np.array(X_test)
print(X_test)
```

```python
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
predicted_value = regressor.predict(X_test)
predicted_value = sc.inverse_transform(predicted_value)
print(predicted_value)
print(len(real_value))
print(len(predicted_value))
#val2=np.argmax(predicted_value,axis=2)
#print(val2)
val0=np.argmax(predicted_value)
print(val0)
```

```python
print("\nMax element : ", max(predicted_value))

if max(predicted_value)>2000:
    print("Flood may Occur")
else:
     print("No flood in next 20 years")
```

```python
# Visualising the results
plt.plot(real_value, color = 'red', label = 'Real Prediction')
plt.plot(predicted_value, color = 'blue', label = 'Proposed Prediction')
plt.title('Flood Prediction')
plt.xlabel('Next Few Years')
plt.ylabel('Rainfall Value')
plt.ylim([0, 10000])
plt.legend()
```

```python
print("")
print('-------------------- Model Summary --------------------')
regressor.summary() # print model summary
print("")
```

```python
from sklearn.metrics import mean_squared_error
print('---------- Evaluation on Test Data ----------')
print("MSE: ", mean_squared_error(predicted_value, real_value)/1000000)
print("")
```

```python
import numpy as np
# Calculate the mean absolute percentage error between predicted and real values
mape = np.mean(np.abs((real_value - predicted_value) / real_value)) * 100

# Calculate the accuracy as 100% minus the mean absolute percentage error
accuracy = 100 - mape

# Print the accuracy
print("Accuracy: {:.2f}%".format(accuracy))
```
